## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df['line_subtotal'] = df['unit_price'] * df['qty']
print(df[['unit_price', 'qty', 'line_subtotal']].head(2))
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
#df['total_weight'] = df['unit_weight'] * df['qty'] #issues with this code, needed to change the order of calulation
#df['shipping_price'] = np.where(df['unit_weight'] > 50, df['unit_weight'] * df['qty'] * 7, df['unit_weight'] *df['qty'] * 10)
df['total_weight'] = df['unit_weight'] * df['qty']
df['shipping_price'] = np.where(df['total_weight'] > 50, df['total_weight'] * 7, df['total_weight'] * 10) # this was a gpt suggestion with the understanding that calculating the total weight first made sure it was available for use in the next line

print(df[['unit_price', 'unit_weight', 'qty', 'total_weight', 'shipping_price']].head(3))
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
df['line_price'] = (df['line_subtotal'] + df['shipping_price']) * 1.0925
print(df[['line_subtotal', 'shipping_price', 'line_price']].head(3).round(2))
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
# there is one less row than the example, 'total_weight' total weight would be used to calculate shipping weight, validating against 'total_price'
df['line_cost'] = (df['unit_cost'] * df['qty']) + df['shipping_price']
df.head(3)
```

```python
# Create a column for the profit of each line using line cost and line price
df['profit'] = df['line_price'] - df['line_cost']
df.head(3)


```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
top_5
for client_id in top_5:
    total_spent = df[df['client_id'] == client_id]['line_price'].sum()
    print(f"Client ID {client_id} spent: ${total_spent:,.2f}")


```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit.
#total cost is not included in the instructions but it is in the example output.  Matching the example output
print(df.columns) 
top_5_df = df[df['client_id'].isin(top_5)]
summary_df = top_5_df.groupby('client_id').agg(
    total_units=('qty', 'sum'),
    total_shipping_price=('shipping_price', 'sum'),
    total_revenue=('line_price', 'sum'),
    total_cost=('line_cost', 'sum'),
    total_profit=('profit', 'sum')
).reset_index()
pd.set_option('display.float_format', '{:.2f}'.format)
summary_df
```

```python
# Format the data and rename the columns to names suitable for presentation.

# Define the money columns. 
money_columns = ['total_shipping_price', 'total_revenue', 'total_cost', 'total_profit']

# Define a function that converts a dollar amount to millions.
def conv_to_millions(value):
    return value / 1_000_000

# Apply the currency_format_millions function to only the money columns. 
summary_df[money_columns] = summary_df[money_columns].apply(conv_to_millions)

# Rename the columns to reflect the change in the money format. 
summary_df.rename(columns={
    'client_id': 'Customer ID',
    'total_shipping_price': 'Shipping (millions)',
    'total_revenue': 'Total Revenue (millions)',
    'total_cost': 'Total Cost (millions)',
    'total_profit': 'Total Profit (millions)',
    'total_units': 'Units'
}, inplace=True)

summary_df
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
sorted_summary_df = summary_df.sort_values(by='Total Profit (millions)', ascending=False).reset_index(drop=True)
sorted_summary_df
```
